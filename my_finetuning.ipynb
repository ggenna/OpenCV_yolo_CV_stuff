```python
import cv2
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset
from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader



from ultralytics import YOLO

def run_hyperparameter_tuning():
    """
    Esegue l'ottimizzazione degli iperparametri (tuning) per
    trovare la configurazione migliore.
    """

    # Carica il modello pre-addestrato
    PRETRAINED_MODEL = 'yolov8m.pt'
    model = YOLO(PRETRAINED_MODEL)
    # Display model information (optional)
    model.info()

    # Define search space
    search_space = {
    "lr0": (1e-5, 1e-1),
    "degrees": (0.0, 45.0),
    }

    # Percorso al tuo file di configurazione del dataset
    DATA_YAML_PATH = '/content/drive/MyDrive/AI/miei-test-deeplearning/HAND_Steering-wheel_phone/driver_dataset.yaml'

    print(f"Avvio Ottimizzazione Iperparametri per: {PRETRAINED_MODEL}")
    print(f"Dataset: {DATA_YAML_PATH}")

    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    try:
        # Avvia il processo di tuning
        results = model.tune(
            data=DATA_YAML_PATH,

            # Quante combinazioni diverse provare.
            # Inizia con un valore basso (es. 30-50) per testare.
            iterations=30,

            # Parametri di base per ogni run di tuning
            epochs=30,         # Meno epoche, perchÃ© testa molte volte
            imgsz=640,
            batch=8,
            device=device,

            # Parametri da non ottimizzare (fissi)
            project='Driver_Detection_Tuning',
            name='yolov8m_auto_tune',

            # Usa uno spazio di ricerca genetico per l'evoluzione
            # Cerca i migliori: lr0, lrf, momentum, weight_decay, ecc.
            optimizer='AdamW', # O 'SGD'. AdamW Ã¨ spesso un buon default.

            # Parametri di augmentation (fissi durante il tuning)
            degrees=10,
            translate=0.1,
            scale=0.2,
            fliplr=0.5,
            mosaic=1.0,
            space=search_space,
            resume=True
        )

        print("\n--- Tuning Completato ---")
        print("I risultati migliori e i parametri consigliati sono stati salvati.")
        print(f"Cartella dei risultati: {results.save_dir}")

        # Ora puoi prendere i parametri migliori e usarli
        # per un training completo (come nell'Opzione A)

        # Stampa i migliori iperparametri trovati
        print(f"Iperparametri migliori trovati: {results.best_hyp}")
        print(f"Fitness migliore: {results.best_fitness}")


    except Exception as e:
        print(f"Errore durante il tuning: {e}")

if __name__ == "__main__":
    # Carica il modello pre-addestrato
    #PRETRAINED_MODEL = 'yolov8m.pt'
    #model = YOLO(PRETRAINED_MODEL)
    # Display model information (optional)
    #model.info()  # Stampa info complete del modello (layer, parametri, ecc.)
    #model.info(verbose=True)  # Versione dettagliata
    # NOMI DELLE CLASSI
    #print(model.names)  # Dizionario {0: 'person', 1: 'bicycle', 2: 'car', ...}

    # INFORMAZIONI SUL MODELLO
    #print(model.task)  # 'detect', 'segment', 'classify', 'pose'
    #print(model.device)  # Device su cui gira il modello
    #print(model.ckpt_path)  # Percorso del checkpoint caricato

    # MODELLO PYTORCH
    #print(model.model)  # Il modello PyTorch sottostante
    #print(model.ckpt)  # Checkpoint completo caricato

    #categories = model.model_name;

    #print("stampa model_name: ",model.model_name)
    #print("stampa get_submodule: ",model.get_submodule)

    #print("stampa categorie: ",categories)
    run_hyperparameter_tuning()
```

```python
import torch
import pickle
import io

def safe_load_pt(model_path):
    """
    Carica un file .pt in modo sicuro senza dipendenze esterne
    """
    print("="*80)
    print(f"ANALISI FILE: {model_path}")
    print("="*80)

    # Dimensione file
    import os
    file_size = os.path.getsize(model_path) / (1024 * 1024)
    print(f"\nğŸ“¦ Dimensione: {file_size:.2f} MB")

    try:
        # Metodo 1: Prova caricamento standard (PyTorch 2.6+)
        print("\nğŸ” Tentativo 1: Caricamento standard...")
        try:
            data = torch.load(model_path, map_location='cpu', weights_only=True)
            print("âœ… Caricato con weights_only=True (solo tensori)")
            return data, "weights_only"
        except Exception as e1:
            print(f"âŒ Fallito: {str(e1)[:100]}")

            # Metodo 2: Caricamento unsafe (necessario per modelli con classi custom)
            print("\nğŸ” Tentativo 2: Caricamento unsafe...")
            try:
                data = torch.load(model_path, map_location='cpu', weights_only=False)
                print("âœ… Caricato con weights_only=False")
                return data, "full"
            except Exception as e2:
                print(f"âŒ Fallito: {str(e2)[:200]}")

                # Metodo 3: Analisi binaria del file
                print("\nğŸ” Tentativo 3: Analisi binaria...")
                return analyze_pickle_binary(model_path), "binary"

    except Exception as e:
        print(f"\nâŒ Errore fatale: {str(e)}")
        return None, "error"

def analyze_pickle_binary(filepath):
    """
    Analizza il contenuto pickle senza eseguirlo completamente
    """
    with open(filepath, 'rb') as f:
        # Leggi i primi byte per identificare il formato
        magic = f.read(8)
        f.seek(0)

        print(f"\nğŸ”¬ Magic bytes: {magic.hex()}")

        # Prova a leggere le operazioni pickle
        try:
            unpickler = pickle.Unpickler(f)

            # Intercetta i GLOBAL (nomi di classi/moduli)
            globals_found = []

            class InspectUnpickler(pickle.Unpickler):
                def find_class(self, module, name):
                    globals_found.append(f"{module}.{name}")
                    # Non caricare realmente, solo registra
                    return None

            f.seek(0)
            inspector = InspectUnpickler(f)

            try:
                inspector.load()
            except:
                pass  # Ci aspettiamo errori, vogliamo solo i globals

            return {'globals': globals_found, 'magic': magic.hex()}

        except Exception as e:
            return {'error': str(e), 'magic': magic.hex()}

def inspect_data(data, load_type):
    """
    Ispeziona i dati caricati
    """
    print("\n" + "="*80)
    print("ğŸ“‹ CONTENUTO DEL FILE")
    print("="*80)

    if data is None:
        print("âŒ Nessun dato caricato")
        return

    if load_type == "binary":
        print(f"\nğŸ”¬ ANALISI BINARIA:")
        if 'globals' in data:
            print(f"\nğŸ“¦ Classi/Moduli trovati ({len(data['globals'])}):")
            for g in data['globals'][:20]:  # Mostra prime 20
                print(f"  - {g}")
            if len(data['globals']) > 20:
                print(f"  ... e altri {len(data['globals']) - 20}")

            # Identifica il tipo di modello
            print(f"\nğŸ¯ IDENTIFICAZIONE:")
            globals_str = ' '.join(data['globals'])

            if 'ultralytics' in globals_str.lower():
                print("  âœ… Probabilmente YOLO (Ultralytics)")
            elif 'yolo' in globals_str.lower():
                print("  âœ… Probabilmente YOLO")
            elif 'detectron' in globals_str.lower():
                print("  âœ… Probabilmente Detectron2")
            elif 'transformers' in globals_str.lower():
                print("  âœ… Probabilmente HuggingFace Transformers")
            elif 'torch.nn' in globals_str:
                print("  âœ… Modello PyTorch generico")
            else:
                print("  â“ Tipo sconosciuto")
        return

    print(f"Tipo oggetto: {type(data)}")

    # Se Ã¨ un dizionario
    if isinstance(data, dict):
        print(f"\nğŸ”‘ Chiavi trovate ({len(data)}):")
        for key in data.keys():
            value_type = type(data[key]).__name__

            # Mostra info aggiuntive per alcuni tipi
            if isinstance(data[key], dict):
                info = f"dict con {len(data[key])} elementi"
            elif isinstance(data[key], (list, tuple)):
                info = f"{value_type} con {len(data[key])} elementi"
            elif hasattr(data[key], 'shape'):
                info = f"Tensor {data[key].shape}"
            else:
                info = value_type

            print(f"  - {key}: {info}")

        # Analisi dettagliata di chiavi comuni
        print(f"\n" + "="*80)
        print("ğŸ“Š ANALISI DETTAGLIATA")
        print("="*80)

        # Epoch
        if 'epoch' in data:
            print(f"ğŸ“ˆ Epoca: {data['epoch']}")

        # Model
        if 'model' in data:
            print(f"\nğŸ¤– MODEL:")
            model = data['model']
            print(f"  Tipo: {type(model).__name__}")

            # Prova ad accedere ad attributi comuni
            for attr in ['nc', 'names', 'yaml', 'stride']:
                if hasattr(model, attr):
                    value = getattr(model, attr)
                    print(f"  {attr}: {value}")

        # State dict
        if 'state_dict' in data:
            state_dict = data['state_dict']
            print(f"\nâš–ï¸ STATE_DICT:")
            print(f"  Numero parametri: {len(state_dict)}")
            print(f"  Prime 5 chiavi:")
            for i, key in enumerate(list(state_dict.keys())[:5]):
                shape = state_dict[key].shape if hasattr(state_dict[key], 'shape') else '?'
                print(f"    {i+1}. {key}: {shape}")

            # Conta parametri totali
            total = sum(p.numel() for p in state_dict.values() if hasattr(p, 'numel'))
            print(f"  Parametri totali: {total:,}")

        # Optimizer
        if 'optimizer' in data:
            print(f"\nâš™ï¸ OPTIMIZER:")
            opt = data['optimizer']
            if isinstance(opt, dict) and 'param_groups' in opt:
                pg = opt['param_groups'][0]
                for key in ['lr', 'momentum', 'weight_decay']:
                    if key in pg:
                        print(f"  {key}: {pg[key]}")

        # Best fitness
        if 'best_fitness' in data:
            print(f"\nğŸ† Best fitness: {data['best_fitness']}")

        # Training args
        if 'train_args' in data:
            print(f"\nğŸ¯ TRAIN_ARGS:")
            args = data['train_args']
            if hasattr(args, '__dict__'):
                args_dict = vars(args)
                for k, v in list(args_dict.items())[:10]:
                    print(f"  {k}: {v}")
            elif isinstance(args, dict):
                for k, v in list(args.items())[:10]:
                    print(f"  {k}: {v}")

    # Se Ã¨ un tensor diretto
    elif torch.is_tensor(data):
        print(f"\nğŸ“Š Tensor diretto:")
        print(f"  Shape: {data.shape}")
        print(f"  Dtype: {data.dtype}")
        print(f"  Device: {data.device}")

    # Se Ã¨ un modello PyTorch
    elif hasattr(data, 'state_dict'):
        print(f"\nğŸ¤– Modulo PyTorch:")
        print(f"  Tipo: {type(data).__name__}")
        state = data.state_dict()
        print(f"  Parametri: {len(state)}")
        total = sum(p.numel() for p in state.values())
        print(f"  Parametri totali: {total:,}")

# MAIN
model_path = '/content/drive/MyDrive/AI/miei-test-deeplearning/last.pt'

data, load_type = safe_load_pt(model_path)
#inspect_data(data, load_type)

print("\n" + "="*80)
print("âœ… ANALISI COMPLETATA")
print("="*80)
```

```python
from ultralytics import YOLO
import torch
import yaml

def run_check_model():
    """
    Esegue il training finale caricando manualmente gli iperparametri
    e filtrando quelli non validi per la versione corrente.
    """

    # --- 1. Configurazione Modello ---
    PRETRAINED_MODEL = 'yolo11x.pt'
    model = YOLO(PRETRAINED_MODEL)

    # --- 2. Percorsi ---
    DATA_YAML_PATH = '/content/drive/MyDrive/AI/miei-test-deeplearning/HAND_Steering-wheel_phone/driver_dataset.yaml'
    TUNED_HYP_PATH = '/content/drive/MyDrive/AI/miei-test-deeplearning/best_hyperparameters.yaml'

    # --- 3. Carica Manualmente gli Iperparametri ---
    print(f"Caricamento iperparametri da: {TUNED_HYP_PATH}")
    with open(TUNED_HYP_PATH, 'r') as f:
        tuned_hyp = yaml.safe_load(f)

    print(f"Iperparametri caricati: {tuned_hyp}")


    # --- 5. Parametri di Training Finale ---
    FINAL_EPOCHS = 100
    BATCH_SIZE = 8
    IMAGE_SIZE = 640

    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Avvio training finale con iperparametri ottimizzati.")

    # --- 6. Avvio Training ---
    try:
        results = model.train(
            data=DATA_YAML_PATH,
            **tuned_hyp,  # Usa gli iperparametri
            epochs=FINAL_EPOCHS,
            imgsz=IMAGE_SIZE,
            batch=BATCH_SIZE,
            device=device,
            project='Driver_Detection_FINAL',
            name=f'final_run_epochs{FINAL_EPOCHS}',
            exist_ok=True,
            patience=10
        )

        print("\n--- Training Finale Completato ---")
        print(f"Il tuo modello migliore si trova in: {results.save_dir}/weights/best.pt")

    except Exception as e:
        print(f"Errore durante il training finale: {e}")

if __name__ == "__main__":
    run_check_model()
```
